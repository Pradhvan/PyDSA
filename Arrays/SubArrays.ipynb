```Python

>>> a = [3,-2,4]

>>> [3,1,5,-2,2,4]
```

* `(start, end)` -> `sum` 

* `(0,0)` -> `3`

* `(0,1)` -> `1`

* `(0,2)` -> `5`

* `(1,1)` -> `-2`

* `(1,2)` -> `2`

* `(2,2)` -> `4`

* Time Complexity: `O(n^3)`

* Time Complexity: `O(n^n)`
