```python
def left_max(a):
    """
    Args:
        [list]a
        
    Output:
        [list]
    """
    left_m = []
    lef_m[0] = a[0]
    for i in range(len(a)):
        number = max(lef_m[i-1], a[i])
        lef_m.append(number)
    return lef_m
```
